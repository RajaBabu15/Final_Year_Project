```python
'''Mark2'''

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# -------------------------------
# Activation Functions
# -------------------------------
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

def sigmoid_deriv(x):
    s = sigmoid(x)
    return s * (1 - s)

# -------------------------------
# Standard ML Neural Network (MLNN)
# -------------------------------
class MLNN:
    def __init__(self, input_dim, hidden_dim, output_dim, learning_rate=0.01, momentum=0.9, noise_std=0.001):
        self.learning_rate = learning_rate
        self.momentum = momentum
        self.noise_std = noise_std

        self.W1 = np.random.randn(input_dim, hidden_dim) * 0.1
        self.W2 = np.random.randn(hidden_dim, output_dim) * 0.1

        self.W1_old = np.copy(self.W1)
        self.W2_old = np.copy(self.W2)

    def forward(self, X):
        self.Z1 = np.dot(X, self.W1)
        self.A1 = sigmoid(self.Z1)
        self.Z2 = np.dot(self.A1, self.W2)
        self.A2 = sigmoid(self.Z2)
        return self.A2

    def backward(self, X, Y, output):
        error_output = Y - output
        delta_output = error_output * sigmoid_deriv(self.Z2)
        error_hidden = np.dot(delta_output, self.W2.T)
        delta_hidden = error_hidden * sigmoid_deriv(self.Z1)

        grad_W2 = np.dot(self.A1.T, delta_output)
        grad_W1 = np.dot(X.T, delta_hidden)
        return grad_W1, grad_W2

    def update_weights(self, grad_W1, grad_W2):
        noise_W1 = np.random.normal(0, self.noise_std, self.W1.shape)
        noise_W2 = np.random.normal(0, self.noise_std, self.W2.shape)
        new_W1 = self.W1 + self.learning_rate * grad_W1 + self.momentum * (self.W1 - self.W1_old) + noise_W1
        new_W2 = self.W2 + self.learning_rate * grad_W2 + self.momentum * (self.W2 - self.W2_old) + noise_W2

        self.W1_old = np.copy(self.W1)
        self.W2_old = np.copy(self.W2)
        self.W1 = new_W1
        self.W2 = new_W2

    def train(self, X, Y, epochs=500):
        losses = []
        for epoch in range(epochs):
            output = self.forward(X)
            grad_W1, grad_W2 = self.backward(X, Y, output)
            self.update_weights(grad_W1, grad_W2)
            loss = np.mean((Y - output) ** 2)
            losses.append(loss)
        return losses

    def predict(self, X):
        return self.forward(X)

# -------------------------------
# Iterative (FIFO-based) Neural Network (IterativeNN)
# -------------------------------
class IterativeNN:
    def __init__(self, input_dim, hidden_dim, output_dim, fifo_size=10, cycles=5, learning_rate=0.01, momentum=0.9, noise_std=0.001):
        self.learning_rate = learning_rate
        self.momentum = momentum
        self.noise_std = noise_std
        self.fifo_size = fifo_size  # np: FIFO size
        self.cycles = cycles        # nc: number of cycles

        self.W1 = np.random.randn(input_dim, hidden_dim) * 0.1
        self.W2 = np.random.randn(hidden_dim, output_dim) * 0.1

        self.W1_old = np.copy(self.W1)
        self.W2_old = np.copy(self.W2)

        self.fifo_stack = []  # FIFO stack for training samples

        # Dummy optimized parameters (for demonstration)
        self.a = 1.2345
        self.b = 0.9876
        self.c = 1.1111
        self.d = 0.5555
        self.e = 0.3333

    def forward(self, X):
        self.Z1 = np.dot(X, self.W1)
        self.A1 = sigmoid(self.Z1)
        self.Z2 = np.dot(self.A1, self.W2)
        self.A2 = sigmoid(self.Z2)
        return self.A2

    def backward(self, X, Y, output):
        error_output = Y - output
        delta_output = error_output * sigmoid_deriv(self.Z2)
        error_hidden = np.dot(delta_output, self.W2.T)
        delta_hidden = error_hidden * sigmoid_deriv(self.Z1)

        grad_W2 = np.dot(self.A1.T, delta_output)
        grad_W1 = np.dot(X.T, delta_hidden)
        return grad_W1, grad_W2

    def update_weights(self, grad_W1, grad_W2):
        noise_W1 = np.random.normal(0, self.noise_std, self.W1.shape)
        noise_W2 = np.random.normal(0, self.noise_std, self.W2.shape)
        new_W1 = self.W1 + self.learning_rate * grad_W1 + self.momentum * (self.W1 - self.W1_old) + noise_W1
        new_W2 = self.W2 + self.learning_rate * grad_W2 + self.momentum * (self.W2 - self.W2_old) + noise_W2

        self.W1_old = np.copy(self.W1)
        self.W2_old = np.copy(self.W2)
        self.W1 = new_W1
        self.W2 = new_W2

    def iterative_train(self, X, Y, iterations=500):
        losses = []
        n_samples = X.shape[0]
        for it in range(iterations):
            # Randomly select one sample
            idx = np.random.randint(0, n_samples)
            x_sample = X[idx:idx+1]
            y_sample = Y[idx:idx+1]

            self.fifo_stack.append((x_sample, y_sample))
            if len(self.fifo_stack) > self.fifo_size:
                self.fifo_stack.pop(0)

            # When FIFO is full, perform training cycles
            if len(self.fifo_stack) == self.fifo_size:
                for _ in range(self.cycles):
                    np.random.shuffle(self.fifo_stack)
                    for (x, y) in self.fifo_stack:
                        output = self.forward(x)
                        grad_W1, grad_W2 = self.backward(x, y, output)
                        self.update_weights(grad_W1, grad_W2)
                # Record average loss over FIFO stack
                cycle_losses = []
                for (x, y) in self.fifo_stack:
                    pred = self.forward(x)
                    cycle_losses.append(np.mean((y - pred) ** 2))
                losses.append(np.mean(cycle_losses))
        return losses

    def predict(self, X):
        return self.forward(X)

# -------------------------------
# Generate Synthetic Data
# -------------------------------
np.random.seed(42)
n_samples = 200
input_dim = 4
X_all = np.random.rand(n_samples, input_dim)

# True relation (simulate kLa): Y = 0.5*x0 + 0.3*x1 + 0.1*x2 - 0.2*x3 + noise
true_weights = np.array([0.5, 0.3, 0.1, -0.2])
Y_all = X_all.dot(true_weights.reshape(-1, 1)) + 0.1 * np.random.randn(n_samples, 1)

# -------------------------------
# Split Data: 80% Train, 20% Test
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(X_all, Y_all, test_size=0.2, random_state=42)

# -------------------------------
# Train MLNN (Batch Model)
# -------------------------------
ml_model = MLNN(input_dim=input_dim, hidden_dim=10, output_dim=1,
                learning_rate=0.01, momentum=0.9, noise_std=0.001)
nn_losses = ml_model.train(X_train, y_train, epochs=500)
nn_train_pred = ml_model.predict(X_train)
nn_test_pred = ml_model.predict(X_test)

# Compute performance metrics for MLNN
nn_train_r2 = r2_score(y_train, nn_train_pred)
nn_test_r2 = r2_score(y_test, nn_test_pred)
nn_train_mse = mean_squared_error(y_train, nn_train_pred)
nn_test_mse = mean_squared_error(y_test, nn_test_pred)

# -------------------------------
# Train IterativeNN (FIFO-based Model)
# -------------------------------
iter_model = IterativeNN(input_dim=input_dim, hidden_dim=10, output_dim=1,
                         fifo_size=10, cycles=5, learning_rate=0.01, momentum=0.9, noise_std=0.001)
iter_losses = iter_model.iterative_train(X_train, y_train, iterations=500)
iter_train_pred = iter_model.predict(X_train)
iter_test_pred = iter_model.predict(X_test)

# Compute performance metrics for IterativeNN
iter_train_r2 = r2_score(y_train, iter_train_pred)
iter_test_r2 = r2_score(y_test, iter_test_pred)
iter_train_mse = mean_squared_error(y_train, iter_train_pred)
iter_test_mse = mean_squared_error(y_test, iter_test_pred)

# -------------------------------
# Print Results and Plot
# -------------------------------
print("\nPerformance Results:")
print(f"Neural Network - Train R: {nn_train_r2:.4f}, Test R: {nn_test_r2:.4f}")
print(f"Neural Network - Train MSE: {nn_train_mse:.6f}, Test MSE: {nn_test_mse:.6f}")
print(f"Iterative Model - Train R: {iter_train_r2:.4f}, Test R: {iter_test_r2:.4f}")
print(f"Iterative Model - Train MSE: {iter_train_mse:.6f}, Test MSE: {iter_test_mse:.6f}")

plt.figure(figsize=(15, 10))

# 1. Plot training loss over epochs/iterations
plt.subplot(2, 2, 1)
plt.plot(range(1, len(nn_losses)+1), nn_losses, 'b-', label='NN Model')
plt.plot(range(1, len(iter_losses)+1), iter_losses, 'r-', label='Iterative Model')
plt.xlabel('Epoch/Iteration')
plt.ylabel('Mean Squared Error')
plt.title('Training Loss vs. Epoch/Iteration')
plt.legend()
plt.grid(True)

# 2. Neural Network: Predicted vs. Actual kLa
plt.subplot(2, 2, 2)
plt.scatter(y_test, nn_test_pred, color='blue', alpha=0.7, label=f"NN Model (R = {nn_test_r2:.4f})")
min_val = min(y_test.min(), nn_test_pred.min())
max_val = max(y_test.max(), nn_test_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', label="Ideal")
plt.xlabel("Actual kLa")
plt.ylabel("Predicted kLa")
plt.title("Neural Network: Predicted vs. Actual kLa")
plt.legend()
plt.grid(True)

# 3. Iterative Model: Predicted vs. Actual kLa
plt.subplot(2, 2, 3)
plt.scatter(y_test, iter_test_pred, color='red', alpha=0.7, label=f"Iterative Model (R = {iter_test_r2:.4f})")
min_val = min(y_test.min(), iter_test_pred.min())
max_val = max(y_test.max(), iter_test_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 'b--', label="Ideal")
plt.xlabel("Actual kLa")
plt.ylabel("Predicted kLa")
plt.title("Iterative Model: Predicted vs. Actual kLa")
plt.legend()
plt.grid(True)

# 4. Combined Comparison on Test Data
plt.subplot(2, 2, 4)
plt.scatter(y_test, nn_test_pred, color='blue', marker='o', alpha=0.7, label=f"NN Model (R = {nn_test_r2:.4f})")
plt.scatter(y_test, iter_test_pred, color='red', marker='x', alpha=0.7, label=f"Iterative Model (R = {iter_test_r2:.4f})")
min_val = min(y_test.min(), min(nn_test_pred.min(), iter_test_pred.min()))
max_val = max(y_test.max(), max(nn_test_pred.max(), iter_test_pred.max()))
plt.plot([min_val, max_val], [min_val, max_val], 'k--', label="Ideal")
plt.xlabel("Actual kLa")
plt.ylabel("Predicted kLa")
plt.title("Comparison: Predicted vs. Actual kLa")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('bubble_column_comparison.png')
plt.show()

# Print the optimized parameters for the iterative model
print("\nOptimized Iterative Model Parameters:")
print(f"a = {iter_model.a:.4f}")
print(f"b (gas velocity exponent) = {iter_model.b:.4f}")
print(f"c (column diameter exponent) = {iter_model.c:.4f}")
print(f"d (liquid height exponent) = {iter_model.d:.4f}")
print(f"e (other factor exponent) = {iter_model.e:.4f}")
```

```python
'''Mark3'''

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# Set random seed for reproducibility
np.random.seed(42)

# ------------- Sigmoid Neural Network (ML Approach) -------------

def sigmoid(x):
    """Sigmoid activation function."""
    return 1.0 / (1.0 + np.exp(-x))

def sigmoid_deriv(x):
    """Derivative of the sigmoid function."""
    s = sigmoid(x)
    return s * (1 - s)

class BubbleColumnNN:
    def __init__(self, input_dim, hidden_dim, output_dim, learning_rate=0.01, momentum=0.9, noise_std=0.001):
        self.learning_rate = learning_rate
        self.momentum = momentum
        self.noise_std = noise_std

        # Initialize weights with small random numbers
        self.W1 = np.random.randn(input_dim, hidden_dim) * 0.1
        self.W2 = np.random.randn(hidden_dim, output_dim) * 0.1

        # Store previous weights for momentum calculation
        self.W1_old = np.copy(self.W1)
        self.W2_old = np.copy(self.W2)

    def forward(self, X):
        """Perform a forward pass."""
        self.Z1 = np.dot(X, self.W1)
        self.A1 = sigmoid(self.Z1)
        self.Z2 = np.dot(self.A1, self.W2)
        self.A2 = sigmoid(self.Z2)
        return self.A2

    def backward(self, X, Y, output):
        """Back-propagate the error and compute gradients."""
        error_output = Y - output
        delta_output = error_output * sigmoid_deriv(self.Z2)

        error_hidden = np.dot(delta_output, self.W2.T)
        delta_hidden = error_hidden * sigmoid_deriv(self.Z1)

        # Calculate gradients
        grad_W2 = np.dot(self.A1.T, delta_output)
        grad_W1 = np.dot(X.T, delta_hidden)
        return grad_W1, grad_W2

    def update_weights(self, grad_W1, grad_W2):
        """Update the weights using a recursive formula with momentum and noise."""
        # Generate small random noise for each weight update
        noise_W1 = np.random.normal(0, self.noise_std, self.W1.shape)
        noise_W2 = np.random.normal(0, self.noise_std, self.W2.shape)

        # Update rule: current weight + learning term + momentum term + noise
        new_W1 = self.W1 + self.learning_rate * grad_W1 + self.momentum * (self.W1 - self.W1_old) + noise_W1
        new_W2 = self.W2 + self.learning_rate * grad_W2 + self.momentum * (self.W2 - self.W2_old) + noise_W2

        # Shift the weights: current become previous for the next update
        self.W1_old = np.copy(self.W1)
        self.W2_old = np.copy(self.W2)
        self.W1 = new_W1
        self.W2 = new_W2

    def train(self, X, Y, epochs=500, verbose=True):
        """Train the neural network and record loss at each epoch."""
        losses = []
        for epoch in range(epochs):
            # Forward pass to get current predictions
            output = self.forward(X)
            # Compute gradients using back-propagation
            grad_W1, grad_W2 = self.backward(X, Y, output)
            # Update weights with the recursive update rule
            self.update_weights(grad_W1, grad_W2)
            # Calculate mean squared error loss
            loss = np.mean((Y - output) ** 2)
            losses.append(loss)
            if verbose and (epoch % 50 == 0 or epoch == epochs-1):
                print(f"Epoch {epoch+1}/{epochs} - Loss: {loss:.6f}")
        return losses

# ------------- Iterative Approach -------------

class IterativeBubbleColumnModel:
    """
    Iterative model for bubble column mass transfer coefficient (kLa) prediction
    using empirical correlation approach.
    """
    def __init__(self):
        # Initialize model parameters for the empirical correlation
        # These will be optimized during training
        self.a = np.random.rand()  # Base coefficient
        self.b = np.random.rand()  # Exponent for gas velocity
        self.c = np.random.rand()  # Exponent for column diameter
        self.d = np.random.rand()  # Exponent for liquid height
        self.e = np.random.rand()  # Additional parameter for other factors

    def predict(self, X):
        """
        Predict kLa using the empirical correlation:
        kLa = a * (gas_velocity)^b * (column_diameter)^c * (liquid_height)^d * (other_factor)^e

        Parameters:
        X: array of shape (n_samples, 4)
           Features representing gas velocity, column diameter, liquid height, and other factors

        Returns:
        Array of predicted kLa values
        """
        gas_velocity = X[:, 0]
        column_diameter = X[:, 1]
        liquid_height = X[:, 2]
        other_factor = X[:, 3]

        # Empirical correlation formula
        kLa = self.a * (gas_velocity ** self.b) * (column_diameter ** self.c) * \
              (liquid_height ** self.d) * (other_factor ** self.e)

        return kLa.reshape(-1, 1)

    def loss(self, X, Y):
        """Calculate mean squared error between predictions and actual values."""
        predictions = self.predict(X)
        return np.mean((Y - predictions) ** 2)

    def gradient(self, X, Y):
        """
        Calculate numerical gradients for model parameters.
        Uses finite difference method.
        """
        epsilon = 1e-5
        grad = np.zeros(5)  # Gradient for a, b, c, d, e

        # Original loss
        original_loss = self.loss(X, Y)

        # Perturb each parameter slightly and measure change in loss
        # Parameter a
        self.a += epsilon
        grad[0] = (self.loss(X, Y) - original_loss) / epsilon
        self.a -= epsilon

        # Parameter b
        self.b += epsilon
        grad[1] = (self.loss(X, Y) - original_loss) / epsilon
        self.b -= epsilon

        # Parameter c
        self.c += epsilon
        grad[2] = (self.loss(X, Y) - original_loss) / epsilon
        self.c -= epsilon

        # Parameter d
        self.d += epsilon
        grad[3] = (self.loss(X, Y) - original_loss) / epsilon
        self.d -= epsilon

        # Parameter e
        self.e += epsilon
        grad[4] = (self.loss(X, Y) - original_loss) / epsilon
        self.e -= epsilon

        return grad

    def train(self, X, Y, epochs=500, learning_rate=0.01, verbose=True):
        """Train the model using gradient descent."""
        losses = []

        for epoch in range(epochs):
            # Calculate gradients
            grad = self.gradient(X, Y)

            # Update parameters with gradient descent
            self.a -= learning_rate * grad[0]
            self.b -= learning_rate * grad[1]
            self.c -= learning_rate * grad[2]
            self.d -= learning_rate * grad[3]
            self.e -= learning_rate * grad[4]

            # Calculate and store loss
            loss = self.loss(X, Y)
            losses.append(loss)

            if verbose and (epoch % 50 == 0 or epoch == epochs-1):
                print(f"Epoch {epoch+1}/{epochs} - Loss: {loss:.6f}")

        return losses

# ------------- Generate dataset -------------

def generate_bubble_column_data(n_samples=100):
    """
    Generate synthetic data for bubble column mass transfer modeling.

    Parameters:
    - n_samples: Number of data points to generate

    Returns:
    - X: Input features (gas velocity, column diameter, liquid height, other factor)
    - Y: Target kLa values
    """
    # Generate input features
    gas_velocity = np.random.uniform(0.01, 0.2, n_samples)  # m/s
    column_diameter = np.random.uniform(0.1, 1.0, n_samples)  # m
    liquid_height = np.random.uniform(0.5, 3.0, n_samples)  # m
    other_factor = np.random.uniform(0.1, 1.0, n_samples)  # dimensionless

    # Combine into feature matrix
    X = np.column_stack((gas_velocity, column_diameter, liquid_height, other_factor))

    # Generate target values using a realistic correlation for kLa
    # Based on simplified version of empirical correlations in literature
    a_true = 0.32
    b_true = 0.7  # Effect of gas velocity
    c_true = -0.2  # Effect of column diameter
    d_true = -0.5  # Effect of liquid height
    e_true = 0.3  # Effect of other factor

    # Calculate kLa using the correlation
    kLa = a_true * (gas_velocity ** b_true) * (column_diameter ** c_true) * \
          (liquid_height ** d_true) * (other_factor ** e_true)

    # Add some random noise
    noise = np.random.normal(0, 0.01, n_samples)
    kLa = kLa + noise

    # Ensure kLa values are positive
    kLa = np.maximum(kLa, 0.001)

    # Reshape to column vector
    Y = kLa.reshape(-1, 1)

    return X, Y

# ------------- Main execution -------------

# Generate more realistic bubble column data
X, Y = generate_bubble_column_data(n_samples=200)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Normalize the input data for the neural network
X_mean = np.mean(X_train, axis=0)
X_std = np.std(X_train, axis=0)
X_train_norm = (X_train - X_mean) / X_std
X_test_norm = (X_test - X_mean) / X_std

# 1. Train the NN model
print("Training Neural Network model...")
nn_model = BubbleColumnNN(input_dim=4, hidden_dim=10, output_dim=1,
                          learning_rate=0.01, momentum=0.9, noise_std=0.001)
nn_losses = nn_model.train(X_train_norm, y_train, epochs=500, verbose=True)

# 2. Train the iterative model
print("\nTraining Iterative model...")
iter_model = IterativeBubbleColumnModel()
iter_losses = iter_model.train(X_train, y_train, epochs=500, learning_rate=0.01, verbose=True)

# Get predictions for both models
nn_train_pred = nn_model.forward(X_train_norm)
nn_test_pred = nn_model.forward(X_test_norm)

iter_train_pred = iter_model.predict(X_train)
iter_test_pred = iter_model.predict(X_test)

# Calculate metrics
nn_train_r2 = r2_score(y_train, nn_train_pred)
nn_test_r2 = r2_score(y_test, nn_test_pred)
nn_train_mse = mean_squared_error(y_train, nn_train_pred)
nn_test_mse = mean_squared_error(y_test, nn_test_pred)

iter_train_r2 = r2_score(y_train, iter_train_pred)
iter_test_r2 = r2_score(y_test, iter_test_pred)
iter_train_mse = mean_squared_error(y_train, iter_train_pred)
iter_test_mse = mean_squared_error(y_test, iter_test_pred)

# Print results
print("\nPerformance Results:")
print(f"Neural Network - Train R: {nn_train_r2:.4f}, Test R: {nn_test_r2:.4f}")
print(f"Neural Network - Train MSE: {nn_train_mse:.6f}, Test MSE: {nn_test_mse:.6f}")
print(f"Iterative Model - Train R: {iter_train_r2:.4f}, Test R: {iter_test_r2:.4f}")
print(f"Iterative Model - Train MSE: {iter_train_mse:.6f}, Test MSE: {iter_test_mse:.6f}")

# Plot the results
plt.figure(figsize=(15, 10))

# 1. Plot training loss over epochs
plt.subplot(2, 2, 1)
plt.plot(range(1, len(nn_losses)+1), nn_losses, 'b-', label='NN Model')
plt.plot(range(1, len(iter_losses)+1), iter_losses, 'r-', label='Iterative Model')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.title('Training Loss vs. Epoch')
plt.legend()
plt.grid(True)

# 2. Neural Network: Predicted vs. Actual kLa
plt.subplot(2, 2, 2)
plt.scatter(y_test, nn_test_pred, color='blue', alpha=0.7, label=f"NN Model (R = {nn_test_r2:.4f})")
# Plot a diagonal line for reference (ideal prediction)
min_val = min(y_test.min(), nn_test_pred.min())
max_val = max(y_test.max(), nn_test_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', label="Ideal")
plt.xlabel("Actual kLa")
plt.ylabel("Predicted kLa")
plt.title("Neural Network: Predicted vs. Actual kLa")
plt.legend()
plt.grid(True)

# 3. Iterative Model: Predicted vs. Actual kLa
plt.subplot(2, 2, 3)
plt.scatter(y_test, iter_test_pred, color='red', alpha=0.7, label=f"Iterative Model (R = {iter_test_r2:.4f})")
# Plot a diagonal line for reference (ideal prediction)
min_val = min(y_test.min(), iter_test_pred.min())
max_val = max(y_test.max(), iter_test_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 'b--', label="Ideal")
plt.xlabel("Actual kLa")
plt.ylabel("Predicted kLa")
plt.title("Iterative Model: Predicted vs. Actual kLa")
plt.legend()
plt.grid(True)

# 4. Combined comparison on test data
plt.subplot(2, 2, 4)
plt.scatter(y_test, nn_test_pred, color='blue', marker='o', alpha=0.7, label=f"NN Model (R = {nn_test_r2:.4f})")
plt.scatter(y_test, iter_test_pred, color='red', marker='x', alpha=0.7, label=f"Iterative Model (R = {iter_test_r2:.4f})")
# Plot a diagonal line for reference (ideal prediction)
min_val = min(y_test.min(), min(nn_test_pred.min(), iter_test_pred.min()))
max_val = max(y_test.max(), max(nn_test_pred.max(), iter_test_pred.max()))
plt.plot([min_val, max_val], [min_val, max_val], 'k--', label="Ideal")
plt.xlabel("Actual kLa")
plt.ylabel("Predicted kLa")
plt.title("Comparison: Predicted vs. Actual kLa")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('bubble_column_comparison.png')  # Save the figure
plt.show()

# Print the optimized parameters for the iterative model
print("\nOptimized Iterative Model Parameters:")
print(f"a = {iter_model.a:.4f}")
print(f"b (gas velocity exponent) = {iter_model.b:.4f}")
print(f"c (column diameter exponent) = {iter_model.c:.4f}")
print(f"d (liquid height exponent) = {iter_model.d:.4f}")
print(f"e (other factor exponent) = {iter_model.e:.4f}")
```

```python
'''Mark4'''


import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# Set random seed for reproducibility
np.random.seed(42)

# ------------- Activation Functions -------------
def sigmoid(x):
    """Sigmoid activation function with numerical stability."""
    # Clip values to avoid overflow
    return 1.0 / (1.0 + np.exp(-np.clip(x, -30, 30)))

def sigmoid_deriv(x):
    """Derivative of the sigmoid function."""
    s = sigmoid(x)
    return s * (1 - s)

# ------------- ML Neural Network -------------
class MLNN:
    def __init__(self, input_dim, hidden_dim, output_dim, learning_rate=0.01, momentum=0.9, noise_std=0.001):
        self.learning_rate = learning_rate
        self.momentum = momentum
        self.noise_std = noise_std

        # Initialize weights with better scaling based on input dimension
        self.W1 = np.random.randn(input_dim, hidden_dim) * np.sqrt(1/input_dim)
        self.W2 = np.random.randn(hidden_dim, output_dim) * np.sqrt(1/hidden_dim)

        # Store previous weights for momentum calculation
        self.W1_old = np.copy(self.W1)
        self.W2_old = np.copy(self.W2)

    def forward(self, X):
        """Perform a forward pass."""
        self.Z1 = np.dot(X, self.W1)
        self.A1 = sigmoid(self.Z1)
        self.Z2 = np.dot(self.A1, self.W2)
        self.A2 = sigmoid(self.Z2)
        return self.A2

    def backward(self, X, Y, output):
        """Back-propagate the error and compute gradients."""
        error_output = Y - output
        delta_output = error_output * sigmoid_deriv(self.Z2)

        error_hidden = np.dot(delta_output, self.W2.T)
        delta_hidden = error_hidden * sigmoid_deriv(self.Z1)

        # Calculate gradients
        grad_W2 = np.dot(self.A1.T, delta_output)
        grad_W1 = np.dot(X.T, delta_hidden)
        return grad_W1, grad_W2

    def update_weights(self, grad_W1, grad_W2):
        """Update the weights using momentum and adaptive noise."""
        # Generate small random noise with adaptive scaling
        noise_W1 = np.random.normal(0, self.noise_std * np.sqrt(np.mean(grad_W1**2)), self.W1.shape)
        noise_W2 = np.random.normal(0, self.noise_std * np.sqrt(np.mean(grad_W2**2)), self.W2.shape)

        # Update rule with gradient clipping for stability
        grad_W1_clipped = np.clip(grad_W1, -1, 1)
        grad_W2_clipped = np.clip(grad_W2, -1, 1)

        new_W1 = self.W1 + self.learning_rate * grad_W1_clipped + self.momentum * (self.W1 - self.W1_old) + noise_W1
        new_W2 = self.W2 + self.learning_rate * grad_W2_clipped + self.momentum * (self.W2 - self.W2_old) + noise_W2

        # Shift the weights: current become previous for the next update
        self.W1_old = np.copy(self.W1)
        self.W2_old = np.copy(self.W2)
        self.W1 = new_W1
        self.W2 = new_W2

    def train(self, X, Y, epochs=500, verbose=True, early_stopping=True, patience=50):
        """Train the neural network with early stopping option."""
        losses = []
        best_loss = float('inf')
        patience_counter = 0

        for epoch in range(epochs):
            # Forward pass to get current predictions
            output = self.forward(X)
            # Compute gradients using back-propagation
            grad_W1, grad_W2 = self.backward(X, Y, output)
            # Update weights with the recursive update rule
            self.update_weights(grad_W1, grad_W2)
            # Calculate mean squared error loss
            loss = np.mean((Y - output) ** 2)
            losses.append(loss)

            if verbose and (epoch % 50 == 0 or epoch == epochs-1):
                print(f"Epoch {epoch+1}/{epochs} - Loss: {loss:.6f}")

            # Early stopping logic
            if early_stopping:
                if loss < best_loss:
                    best_loss = loss
                    patience_counter = 0
                else:
                    patience_counter += 1

                if patience_counter >= patience:
                    print(f"Early stopping at epoch {epoch+1}")
                    break

        return losses

    def predict(self, X):
        """Generate predictions for input data."""
        return self.forward(X)

# ------------- Iterative (FIFO-based) Neural Network -------------
class IterativeNN:
    def __init__(self, input_dim, hidden_dim, output_dim, fifo_size=10, cycles=5,
                 learning_rate=0.01, momentum=0.9, noise_std=0.001):
        self.learning_rate = learning_rate
        self.momentum = momentum
        self.noise_std = noise_std
        self.fifo_size = fifo_size  # Size of the FIFO buffer
        self.cycles = cycles        # Number of cycles per iteration

        # Initialize weights with better scaling
        self.W1 = np.random.randn(input_dim, hidden_dim) * np.sqrt(1/input_dim)
        self.W2 = np.random.randn(hidden_dim, output_dim) * np.sqrt(1/hidden_dim)

        self.W1_old = np.copy(self.W1)
        self.W2_old = np.copy(self.W2)

        # FIFO buffer for training samples
        self.fifo_stack = []

        # Initialize physical model parameters (will be optimized)
        self.a = np.random.uniform(0.1, 2.0)
        self.b = np.random.uniform(0.1, 2.0)
        self.c = np.random.uniform(0.1, 2.0)
        self.d = np.random.uniform(0.1, 2.0)
        self.e = np.random.uniform(0.1, 2.0)

    def forward(self, X):
        """Perform a forward pass."""
        self.Z1 = np.dot(X, self.W1)
        self.A1 = sigmoid(self.Z1)
        self.Z2 = np.dot(self.A1, self.W2)
        self.A2 = sigmoid(self.Z2)
        return self.A2

    def backward(self, X, Y, output):
        """Back-propagate the error and compute gradients."""
        error_output = Y - output
        delta_output = error_output * sigmoid_deriv(self.Z2)

        error_hidden = np.dot(delta_output, self.W2.T)
        delta_hidden = error_hidden * sigmoid_deriv(self.Z1)

        # Calculate gradients
        grad_W2 = np.dot(self.A1.T, delta_output)
        grad_W1 = np.dot(X.T, delta_hidden)
        return grad_W1, grad_W2

    def update_weights(self, grad_W1, grad_W2):
        """Update weights with adaptive learning rate."""
        # Generate noise with decaying amplitude based on gradient magnitude
        noise_scale = self.noise_std * np.exp(-0.01 * (np.mean(grad_W1**2) + np.mean(grad_W2**2)))
        noise_W1 = np.random.normal(0, noise_scale, self.W1.shape)
        noise_W2 = np.random.normal(0, noise_scale, self.W2.shape)

        # Gradient clipping for stability
        grad_W1_clipped = np.clip(grad_W1, -1, 1)
        grad_W2_clipped = np.clip(grad_W2, -1, 1)

        # Update rule with momentum and adaptive noise
        new_W1 = self.W1 + self.learning_rate * grad_W1_clipped + self.momentum * (self.W1 - self.W1_old) + noise_W1
        new_W2 = self.W2 + self.learning_rate * grad_W2_clipped + self.momentum * (self.W2 - self.W2_old) + noise_W2

        # Shift the weights
        self.W1_old = np.copy(self.W1)
        self.W2_old = np.copy(self.W2)
        self.W1 = new_W1
        self.W2 = new_W2

    def update_physical_params(self, X, Y, lr=0.001):
        """Update physical model parameters using gradient descent."""
        # Generate predictions with current physical parameters
        predictions = self.physical_predict(X)
        error = Y - predictions

        # Numerical gradient calculation (simplified)
        epsilon = 1e-5
        params = [self.a, self.b, self.c, self.d, self.e]
        gradients = []

        # Calculate gradient for each parameter
        for i in range(len(params)):
            # Temporary parameter changes
            params_plus = params.copy()
            params_plus[i] += epsilon

            # Assign temporary parameters
            self.a, self.b, self.c, self.d, self.e = params_plus
            pred_plus = self.physical_predict(X)

            # Restore original parameters
            self.a, self.b, self.c, self.d, self.e = params

            # Calculate numerical gradient
            grad = np.mean(-(pred_plus - predictions) * error / epsilon)
            gradients.append(grad)

        # Update parameters
        self.a -= lr * gradients[0]
        self.b -= lr * gradients[1]
        self.c -= lr * gradients[2]
        self.d -= lr * gradients[3]
        self.e -= lr * gradients[4]

    def physical_predict(self, X):
        """Predict using physical parameter model."""
        # Extract features and clip them to be non-negative
        f1 = np.clip(X[:, 0], 0, None)  # e.g., gas velocity
        f2 = np.clip(X[:, 1], 0, None)  # e.g., column diameter
        f3 = np.clip(X[:, 2], 0, None)  # e.g., liquid height
        f4 = np.clip(X[:, 3], 0, None)  # e.g., other factors

        # Physical model formula using power-law correlation
        predictions = (self.a *
                      np.power(f1 + 1e-10, self.b) *
                      np.power(f2 + 1e-10, self.c) *
                      np.power(f3 + 1e-10, self.d) *
                      np.power(f4 + 1e-10, self.e))

        return predictions.reshape(-1, 1)

    def iterative_train(self, X, Y, iterations=500, verbose=True):
        """Train using FIFO-based iterative approach with physical parameter updates."""
        losses = []
        physical_losses = []
        n_samples = X.shape[0]

        # Pre-fill the FIFO stack with a batch of samples
        init_indices = np.random.choice(n_samples, min(self.fifo_size, n_samples), replace=False)
        for idx in init_indices:
            self.fifo_stack.append((X[idx:idx+1], Y[idx:idx+1]))

        for it in range(iterations):
            # Randomly select one sample
            idx = np.random.randint(0, n_samples)
            x_sample = X[idx:idx+1]
            y_sample = Y[idx:idx+1]

            # Update FIFO stack
            self.fifo_stack.append((x_sample, y_sample))
            if len(self.fifo_stack) > self.fifo_size:
                self.fifo_stack.pop(0)

            # When FIFO has enough samples, perform training cycles
            if len(self.fifo_stack) >= self.fifo_size // 2:
                # Train neural network
                for _ in range(self.cycles):
                    np.random.shuffle(self.fifo_stack)
                    for (x, y) in self.fifo_stack:
                        output = self.forward(x)
                        grad_W1, grad_W2 = self.backward(x, y, output)
                        self.update_weights(grad_W1, grad_W2)

                # Update physical parameters every few iterations
                if it % 5 == 0:
                    # Use a batch for physical parameter update
                    batch_indices = np.random.choice(n_samples, min(32, n_samples), replace=False)
                    self.update_physical_params(X[batch_indices], Y[batch_indices])

                # Record losses
                if it % 10 == 0 or it == iterations - 1:
                    # Calculate NN loss
                    nn_preds = self.forward(X)
                    nn_loss = np.mean((Y - nn_preds) ** 2)

                    # Calculate physical model loss
                    phys_preds = self.physical_predict(X)
                    phys_loss = np.mean((Y - phys_preds) ** 2)

                    losses.append(nn_loss)
                    physical_losses.append(phys_loss)

                    if verbose and (it % 100 == 0 or it == iterations - 1):
                        print(f"Iteration {it+1}/{iterations} - NN Loss: {nn_loss:.6f}, Physical Loss: {phys_loss:.6f}")

        return losses, physical_losses

    def predict(self, X, mode='nn'):
        """Generate predictions using either neural network or physical model."""
        if mode == 'nn':
            return self.forward(X)
        elif mode == 'physical':
            return self.physical_predict(X)
        elif mode == 'ensemble':
            # Weighted combination of both models
            nn_pred = self.forward(X)
            phys_pred = self.physical_predict(X)
            return 0.7 * nn_pred + 0.3 * phys_pred
        else:
            raise ValueError("Mode must be 'nn', 'physical', or 'ensemble'")

# ------------- Generate realistic bubble column data -------------
def generate_bubble_column_data(n_samples=200):
    """Generate synthetic data for bubble column mass transfer modeling."""
    # Generate input features with realistic ranges
    gas_velocity = np.random.uniform(0.01, 0.2, n_samples)  # m/s
    column_diameter = np.random.uniform(0.1, 1.0, n_samples)  # m
    liquid_height = np.random.uniform(0.5, 3.0, n_samples)  # m
    other_factor = np.random.uniform(0.1, 1.0, n_samples)  # dimensionless

    # Combine into feature matrix
    X = np.column_stack((gas_velocity, column_diameter, liquid_height, other_factor))

    # Generate target values using a typical power-law correlation for kLa
    a_true = 0.32
    b_true = 0.7   # Exponent for gas velocity (typically 0.7-0.9)
    c_true = -0.2  # Exponent for column diameter (typically negative)
    d_true = -0.5  # Exponent for liquid height (typically negative)
    e_true = 0.3   # Exponent for other factors

    # Calculate kLa using the correlation
    kLa = a_true * (gas_velocity ** b_true) * (column_diameter ** c_true) * \
          (liquid_height ** d_true) * (other_factor ** e_true)

    # Add some random noise
    noise = np.random.normal(0, 0.01 * np.mean(kLa), n_samples)
    kLa = kLa + noise

    # Ensure kLa values are positive
    kLa = np.maximum(kLa, 0.001)

    # Scale kLa values to be between 0 and 1 for better training
    kLa_scaled = (kLa - np.min(kLa)) / (np.max(kLa) - np.min(kLa))

    # Reshape to column vector
    Y = kLa_scaled.reshape(-1, 1)

    return X, Y

# ------------- Main execution -------------
# Generate more realistic bubble column data
X, Y = generate_bubble_column_data(n_samples=200)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Scale input features for better numerical stability
X_mean = np.mean(X_train, axis=0)
X_std = np.std(X_train, axis=0)
X_train_norm = (X_train - X_mean) / X_std
X_test_norm = (X_test - X_mean) / X_std

print("Training Neural Network (ML) model...")
ml_model = MLNN(
    input_dim=4,
    hidden_dim=10,
    output_dim=1,
    learning_rate=0.01,
    momentum=0.9,
    noise_std=0.001
)
nn_losses = ml_model.train(X_train_norm, y_train, epochs=500)

print("\nTraining Iterative model...")
iter_model = IterativeNN(
    input_dim=4,
    hidden_dim=10,
    output_dim=1,
    fifo_size=20,  # Increased from 10 to 20 for better stability
    cycles=5,
    learning_rate=0.01,
    momentum=0.9,
    noise_std=0.001
)
iter_losses, phys_losses = iter_model.iterative_train(X_train_norm, y_train, iterations=500)

# Get predictions for both models on training and test sets
nn_train_pred = ml_model.predict(X_train_norm)
nn_test_pred = ml_model.predict(X_test_norm)

# Get predictions for iterative model (using both neural network and ensemble approach)
iter_train_pred = iter_model.predict(X_train_norm, mode='nn')
iter_test_pred = iter_model.predict(X_test_norm, mode='nn')
iter_train_ensemble = iter_model.predict(X_train_norm, mode='ensemble')
iter_test_ensemble = iter_model.predict(X_test_norm, mode='ensemble')

# Calculate metrics for ML model
nn_train_r2 = r2_score(y_train, nn_train_pred)
nn_test_r2 = r2_score(y_test, nn_test_pred)
nn_train_mse = mean_squared_error(y_train, nn_train_pred)
nn_test_mse = mean_squared_error(y_test, nn_test_pred)

# Calculate metrics for iterative model (NN only)
iter_train_r2 = r2_score(y_train, iter_train_pred)
iter_test_r2 = r2_score(y_test, iter_test_pred)
iter_train_mse = mean_squared_error(y_train, iter_train_pred)
iter_test_mse = mean_squared_error(y_test, iter_test_pred)

# Calculate metrics for iterative model (ensemble)
ensemble_train_r2 = r2_score(y_train, iter_train_ensemble)
ensemble_test_r2 = r2_score(y_test, iter_test_ensemble)
ensemble_train_mse = mean_squared_error(y_train, iter_train_ensemble)
ensemble_test_mse = mean_squared_error(y_test, iter_test_ensemble)

# Print results
print("\nPerformance Results:")
print(f"Neural Network - Train R: {nn_train_r2:.4f}, Test R: {nn_test_r2:.4f}")
print(f"Neural Network - Train MSE: {nn_train_mse:.6f}, Test MSE: {nn_test_mse:.6f}")
print(f"Iterative Model (NN) - Train R: {iter_train_r2:.4f}, Test R: {iter_test_r2:.4f}")
print(f"Iterative Model (NN) - Train MSE: {iter_train_mse:.6f}, Test MSE: {iter_test_mse:.6f}")
print(f"Iterative Model (Ensemble) - Train R: {ensemble_train_r2:.4f}, Test R: {ensemble_test_r2:.4f}")
print(f"Iterative Model (Ensemble) - Train MSE: {ensemble_train_mse:.6f}, Test MSE: {ensemble_test_mse:.6f}")

# Plot the results
plt.figure(figsize=(16, 12))

# 1. Plot training loss over epochs
plt.subplot(2, 2, 1)
plt.plot(range(1, len(nn_losses)+1), nn_losses, 'b-', label='ML Model')
plt.plot(range(1, len(iter_losses)+1), iter_losses, 'r-', label='Iterative NN')
plt.plot(range(1, len(phys_losses)+1), phys_losses, 'g-', label='Physical Model')
plt.xlabel('Epoch/Iteration')
plt.ylabel('Mean Squared Error')
plt.title('Training Loss vs. Epoch')
plt.legend()
plt.grid(True)

# 2. Neural Network: Predicted vs. Actual kLa
plt.subplot(2, 2, 2)
plt.scatter(y_test, nn_test_pred, color='blue', alpha=0.7, label=f"ML Model (R = {nn_test_r2:.4f})")
# Plot a diagonal line for reference (ideal prediction)
min_val = min(y_test.min(), nn_test_pred.min())
max_val = max(y_test.max(), nn_test_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', label="Ideal")
plt.xlabel("Actual kLa")
plt.ylabel("Predicted kLa")
plt.title("Neural Network: Predicted vs. Actual kLa")
plt.legend()
plt.grid(True)

# 3. Iterative Model: Predicted vs. Actual kLa
plt.subplot(2, 2, 3)
plt.scatter(y_test, iter_test_pred, color='red', alpha=0.7, label=f"Iterative NN (R = {iter_test_r2:.4f})")
plt.scatter(y_test, iter_test_ensemble, color='green', alpha=0.7, marker='x',
            label=f"Ensemble (R = {ensemble_test_r2:.4f})")
# Plot a diagonal line for reference (ideal prediction)
min_val = min(y_test.min(), min(iter_test_pred.min(), iter_test_ensemble.min()))
max_val = max(y_test.max(), max(iter_test_pred.max(), iter_test_ensemble.max()))
plt.plot([min_val, max_val], [min_val, max_val], 'b--', label="Ideal")
plt.xlabel("Actual kLa")
plt.ylabel("Predicted kLa")
plt.title("Iterative Models: Predicted vs. Actual kLa")
plt.legend()
plt.grid(True)

# 4. Combined comparison on test data
plt.subplot(2, 2, 4)
plt.scatter(y_test, nn_test_pred, color='blue', marker='o', alpha=0.7, label=f"ML Model (R = {nn_test_r2:.4f})")
plt.scatter(y_test, iter_test_ensemble, color='green', marker='x', alpha=0.7,
            label=f"Iterative Ensemble (R = {ensemble_test_r2:.4f})")
# Plot a diagonal line for reference (ideal prediction)
min_val = min(y_test.min(), min(nn_test_pred.min(), iter_test_ensemble.min()))
max_val = max(y_test.max(), max(nn_test_pred.max(), iter_test_ensemble.max()))
plt.plot([min_val, max_val], [min_val, max_val], 'k--', label="Ideal")
plt.xlabel("Actual kLa")
plt.ylabel("Predicted kLa")
plt.title("Comparison: Predicted vs. Actual kLa")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('improved_bubble_column_comparison.png')
plt.show()

# Print the optimized parameters for the iterative model
print("\nOptimized Physical Model Parameters:")
print(f"a = {iter_model.a:.4f}")
print(f"b (gas velocity exponent) = {iter_model.b:.4f}")
print(f"c (column diameter exponent) = {iter_model.c:.4f}")
print(f"d (liquid height exponent) = {iter_model.d:.4f}")
print(f"e (other factor exponent) = {iter_model.e:.4f}")
```
